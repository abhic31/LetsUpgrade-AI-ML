```python
import pandas as pd 
```

```python
dataset = pd.read_csv("general_data.csv")
dataset.head()
```

```python
dataset["Attrition"]=dataset["Attrition"].replace({'No':0,"Yes":1})
```

```python
dataset
```

```python
Y = dataset.Attrition
```

```python
X = dataset[['Age' , 'DistanceFromHome' , 'Education' , 'EmployeeCount']]
```

```python
import statsmodels.api as sm
```

```python
X1 = sm.add_constant(X)
```

```python
logistic = sm.Logit(Y,X1)
```

```python
result = logistic.fit()
```

```python
result.summary()
```
